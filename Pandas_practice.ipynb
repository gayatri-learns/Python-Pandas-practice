```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
#Rename columns
df.rename(columns = {'SALES':'TOTAL_SALES'},inplace = True)
```

```python
#select rows
df.loc[0]
df.iloc[0:5]
```

```python
#Conditional filtering
df[df['COUNTRY'] == 'USA']
df[df['TOTAL_SALES'] > 5000]
df[(df['YEAR_ID'] == 2004) & (df['DEALSIZE'] == 'Large')] 
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Fill missing values:
df['STATE'].fillna('Unknown',inplace = True)
```

```python
df['POSTALCODE'].fillna(df['POSTALCODE'].mode()[0],inplace=True)
```

```python
#drop missing values:
df.dropna()
df.dropna(subset=['STATE'])
```

```python
#Sorting
df.sort_values(by = "TOTAL_SALES",ascending = False)
df.sort_values(by = ['YEAR_ID','TOTAL_SALES'])
```

```python
#Ranking
df['SALES_RANK'] = df['TOTAL_SALES'].rank(ascending=False)
```

```python
#GroupBy Operations
#Total sales by country
df.groupby('COUNTRY')['TOTAL_SALES'].sum()
```

```python
#sales by year
df.groupby('YEAR_ID')['TOTAL_SALES'].sum()
```

```python
#sales by deal size 
df.groupby('DEALSIZE')['TOTAL_SALES'].mean()
```

```python
#Multiple Aggregations
df.groupby('COUNTRY').agg({
    'TOTAL_SALES' : ['sum','mean','max'],
    'QUANTITYORDERED':'sum'
})
```
